```python
import numpy as np

import torch
```

```python
X = np.array([160, 170, 180, 190])

y = np.array([0, 0, 1, 1])
```

```python
X_mean = np.mean(X)
X_std = np.std(X)

X_norm = (X - X_mean) / X_std
```

```python
X_norm_tensor = torch.tensor(X_norm, dtype = torch.float32)

y_tensor = torch.tensor(y, dtype = torch.float32)
```

```python
X_norm_tensor = X_norm_tensor.reshape([-1, 1])

X_norm_tensor
```

```python
y_tensor = y_tensor.reshape([-1, 1])

y_tensor
```

```python
class PerceptronModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = torch.nn.Linear(1, 1)
        
        
    def forward(self, x):
        H = self.linear(x)
        
        return H
```

```python
torch.manual_seed(42)
```

```python
model = PerceptronModel()
```

```python
model.linear.weight
```

```python
model.linear.bias
```

```python
model
```

```python
with torch.no_grad():
    H_test = model(X_norm_tensor)
    
print(f'H_test = {H_test}')
```

```python
learning_rate = 0.1
```

```python
epochs = 1000
```

```python
criterion = torch.nn.BCEWithLogitsLoss()
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python
for epoch in range(epochs):
    optimizer.zero_grad()
    
    H = model(X_norm_tensor)
    
    mean_cost = criterion(H, y_tensor)
    
    mean_cost.backward()
    
    optimizer.step()
    
    if epoch % 100 == 0 or epoch == epochs - 1:
        print(
            f'epoch = {epoch},  '
            f'cost = {mean_cost.item()}, '
            f'weight(a) = {model.linear.weight.item()}, '
            f'bias(b) = {model.linear.bias.item()}'
            )
    
    if epoch < 3:
        print(f'model.linear.weight.grad = {model.linear.weight.grad.item()}, '
              f'model.linear.bias.grad = {model.linear.bias.grad.item()}'
        )
```
