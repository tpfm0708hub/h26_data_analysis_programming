```python
import pandas as pd

df = pd.DataFrame({
    'temperature': [10, 7, 4, 0],
    'sales'      : [30, 50, 60, 80]
})

display(df)
```

```python
#   입력 값
#df['temperature'].values  #     1차원 배열로 출력
#df[['temperature']].values#     2차원 배열로 출력(정수형)
X = df[['temperature']].values.astype('float32')

X
```

```python
#   실제 값
y = df[['sales']].values.astype('float32')

y
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
X
```

```python
y
```

```python
"""
tensorflow Dense(): 선형 회귀를 계산
    layer.Dense(#, ): 진짜 값(y)의 칸 수(#칸)
    input_shape = (#, ): 입력값(x)의 칸, 줄 수(줄은 생략 가능)[X: #칸 _줄]
"""

model = keras.Sequential([
    layers.Dense(1, input_shape = (1, ))
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer = keras.optimizers.SGD(learning_rate = 0.001),
    loss = 'mse'
)
```

```python
history = model.fit(X, y, epochs = 10000, verbose = 0)
```

```python
#history.history['loss']#    cost 변화치 확인
history.history['loss'][-1]
```

```python
weight, bias = model.layers[0].get_weights()

a_value = weight[0][0]
b_value = bias[0]

print(f'weight = {weight[0][0]}')
print(f'bias = {bias[0]}')
#print(f'weight = {weight}')#    2차원 배열
#print(f'bias = {bias}')    #    1차원 배열
```
