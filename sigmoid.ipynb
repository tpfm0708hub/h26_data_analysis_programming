```python
import numpy as np
from math import e
```

```python
X = np.array([160, 170, 180, 190])

X
```

```python
y = np.array([0, 0, 1, 1])

y
```

```python
a = 0.1
b = -17
```

```python
H = a * X + b

H
```

```python
def sigmoid(H):
    return 1 / (1 + e ** (-H))
```

```python
z = sigmoid(H)

z
```

```python
costs = - (y * np.log(z) + (1 - y) * np.log(1 - z))

costs
```

```python
np.mean(costs)
```
