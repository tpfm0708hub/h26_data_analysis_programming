```python
import numpy as np
from math import e
```

```python
X = np.array([160, 170, 180, 190])

X
```

```python
y = np.array([0, 0, 1, 1])

y
```

```python
X_mean = np.mean(X)

X_mean
```

```python
X_std = np.std(X)

X_std
```

```python
X_norm = (X - X_mean) / X_std

X_norm
```

```python
a = 0.1
b = 0.0
```

```python
def sigmoid(H):
    return 1 / (1 + e ** (-H))
```

```python
H_X = (X * a) + b

H_X
```

```python
Z_X = sigmoid(H_X)

Z_X
```

```python
a_미분값 = np.mean( (Z_X - y) * X )

a_미분값
```

```python
epochs = 1000
epsilon = 1e-7
```

```python
epsilon
```

```python
1 - epsilon
```

```python
learning_rate = 0.1
epochs = 1000
```

```python
for epoch in range(1000):
    H = (a * X_norm) + b
    print(f'H = {H}')
    z = sigmoid(H)
    print(f'z = {z}')
    
    grad_a = np.mean((z - y) * X_norm )
    print(f'grad_a = {grad_a}')
    
    grad_b = np.mean(z - y)
    print(f'grad_b = {grad_b}')
    
    a = a - (learning_rate * grad_a)
    b = b - (learning_rate * grad_b)
    
    print(f'update a = {a}, update b = {b}')
    
    # z가 epsilon 보다 작으면 log(0) -> -∞ epsilon 리턴 -> 1e-7
    # z가 epsilon 보다 크면 -log(0) -> ∞ 1 - epslion 리턴 -> 0.9999 리턴
    z_safe = np.clip(z, epsilon, 1 - epsilon)
    costs = -(y * np.log(z_safe) + (1 - y) * np.log(1 - z_safe))
    #costs = -(y * np.log(z) + (1 - y) * np.log(1 - z))
    mean_cost = np.mean(costs)
    print(f'mean_cost = {mean_cost}')
    print('='*30)
```
