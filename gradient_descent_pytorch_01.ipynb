```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'temperature': [10, 7, 4, 0],
    'sales':       [30, 50, 60, 80]
})

display(df)
```

```python
df['temperature'].values
```

```python
import torch
x_tensor = torch.tensor(df['temperature'].values, dtype = torch.float32)

x_tensor
```

```python
y_tensor = torch.tensor(df['sales'].values, dtype = torch.float32)

y_tensor
```

```python
a = torch.tensor(-4.0, requires_grad = True)

a
```

```python
b = torch.tensor(80.0, requires_grad = True)

b
```

```python
learning_rate = 0.001
epochs = 1000
cost_history = []
```

```python
a_initial = a.item()

a_initial
```

```python
b_initial = b.item()

b_initial
```

```python
with torch.no_grad():
    #   진짜값 - 예측값
    error = (y_tensor - (x_tensor * a + b))
    print(f'error = {error}')
    
    initial_cost = torch.mean(error ** 2).item()
    print(f'initial_cost = {initial_cost}')
```

```python
for epoch in range(epochs):
    pred = a * x_tensor + b
    
    error = y_tensor - pred
    
    cost = torch.mean(error ** 2)
    cost.backward()
    
    with torch.no_grad():
        a -= learning_rate * a.grad
        b -= learning_rate * b.grad
        
    a.grad.zero_()
    b.grad.zero_()
    
    cost_history.append(cost.item())
    
    if epoch % 100 == 0:
        print(f'epoch = {epoch}, Cost = {cost}, a = {a}, b = {b}')
```

```python
def run_gradient_descent(learning_rate, epochs):
    print(f'learning_rate = {learning_rate}')    
    a = torch.tensor(-4.0, requires_grad = True)
    b = torch.tensor(80.0, requires_grad = True)    
    for epoch in range(epochs):
        pred = a * x_tensor + b
        error = y_tensor - pred
        cost = torch.mean(error ** 2)
        
        cost.backward()
        
        with torch.no_grad():
            a -= learning_rate * a.grad
            b -= learning_rate * b.grad
        
        a.grad.zero_()
        b.grad.zero_()
        
        #print(f'cost = {cost}, a = {a}, b = {b}, epoch = {epoch}')
        if cost.item()> 100000:
            print('Cost가 너무 큽니다.')
            print(f'cost = {cost}, a = {a}, b = {b}, epoch = {epoch}')
        
        if (epoch + 1) % 100 == 0:
            print(f'Cost = {cost}, a = {a}, b = {b}, epoch = {epoch}')
```

```python
run_gradient_descent(0.1, 1000)# learning_rate가 너무 커서 Cost 최저점 못 찾음 → 발산
```

```python
run_gradient_descent(0.000001, 500000)
```

```python
run_gradient_descent(0.001, 1000)
```
