```python
import numpy as np
import torch
from math import e
```

```python
X = np.array([160, 170, 180, 190])

y = np.array([0,   0,    1,    1])
```

```python
X_mean = np.mean(X)
```

```python
X_std = np.std(X)
```

```python
X_norm = (X - X_mean) / X_std
X_norm
```

```python
X_norm_tensor = torch.tensor(X_norm, dtype = torch.float32)

X_norm_tensor
```

```python
y_tensor = torch.tensor(y, dtype = torch.float32)

y_tensor
```

```python
a = torch.tensor(0.1, dtype = torch.float32, requires_grad = True)

b = torch.tensor(0.0, dtype = torch.float32, requires_grad = True)
```

```python
a
```

```python
b
```

```python
learning_rate = 0.1
epochs = 1000
```

```python
optimizer = torch.optim.SGD([a, b], lr = learning_rate)

optimizer
```

```python
epsilon = 1e-7
```

```python
epoches = 1000

for epoch in range(epoches):
    optimizer.zero_grad()
    
    H = (a * X_norm_tensor) + b
    z = torch.sigmoid(H)
    
    z_safe = torch.clamp(z, epsilon, 1 - epsilon)
    
    costs = -((y_tensor * torch.log(z_safe)) + ((1 - y_tensor) * torch.log(1 - z_safe)))
    mean_cost = torch.mean(costs)
    mean_cost.backward()
    optimizer.step()
    
    if epoch % 100 == 0 or epoch == epochs - 1:
        print(
            f'epoch = {epoch},  '
            f'Cost  = {mean_cost.item()},   '
            f'a     = {a.item()},   '
            f'b     = {b.item()}'
        )
    if epoch < 3:
        grad_a = a.grad.item()
        grad_b = b.grad.item()
        
        print(f' (확인용) a.grad={grad_a}, b.grad={grad_b}')
```
