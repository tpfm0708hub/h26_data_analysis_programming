```python
#   math는 파이썬에서 수학적인 계산을 할 수 있도록 도와주는 라이브러리입니다.
#   여기에서 자연상수 e를 가져옵니다. e는 약 2.71828 값을 가지는 수학적으로 중요한 상수입니다.
from math import e
```

```python
"""
시그모이드(Sigmoid 함수 설명):
    시그모이드 함수의 수식은 다음과 같습니다:
    sigmoid(H(x)) = 1 / (1 + e^(-H(x)))
    시그모이드 함수는 인종기능에서 입력값을 받아서 0과 1사이의 값을 출력하는 함수로,
    입력값을 확률(0~1사이의 값)로 바꿔주는 데 자주 사용됩니다.
"""
```

```python
#   예시 1: 선형 회귀 결과 (H(X))가 -5일 때 시그모이드 값을 계산합니다.
sigmoid_minus_5 = 1 / (1 + e** - (-5))
print(sigmoid_minus_5)

#   계산 결과는 약 0.0067 입니다.
#   선형 회귀 결과 (H(X))가 매우 작을 때(큰 음수일 때), 시그모이드 값은 거의 0에 가깝습니다.
```

```python
#   예시 2: 선형 회귀 결과 (H(X))가 -2.5일 때 시그모이드 값을 계선헙나더,
sigmoid_minus_2_5 = 1 / ( 1 + e ** -(-2.5) )
print(sigmoid_minus_2_5)

#   계산 결과눈 약 0.0759 입니다.
#   선형 회귀 결과 (H(X))가 음수일 때, 시그모이드 값은 0과 0.5 사이의 값이 나옵니다.
```

```python
#   예시 3: 선형 회귀 결과 (H(X))가 0일 때 시그모이드 값을 계산합니다.
sigmoid_0 = 1 / (1 + e ** -(0))
print(sigmoid_0)

#   계산 결과는 정확히 0.5 입니다.
#   선형 회귀 결과 (H(X))가 0일 때 시그모이드 값은 항상 0.5입니다.
```

```python
#   예시 4: 선형 회귀 결과 (H(X))가 2.5일 때 시그모이드 값을 계산합니다.
sigmoid_2_5 = 1 / ( 1 + e ** -(2.5) )
print(sigmoid_2_5)

#   계산 결과는 약 0.9241 입니다.
#   선형 회귀 결과 (H(X))가 양수일 때, 시그모이드 값은 0.5와 1 사이의 값이 나옵니다.
```

```python
#   예시 5: 선형 회귀 결과 (H(X))가 5일 때 시그모이드 값을 계산합니다.
sigmoid_5 = 1 / ( 1 + e ** -(5) )
print(sigmoid_5)

#   계산 결과는 약 0.9933 입니다.
#   선형 회귀 결과 (H(X))가 매우 클 때(큰 양수일 때), 시그모이드 값ㅇ느 거의 1에 가깝습니다.
```

```python
# math 라이브러리에서 자연상수 e를 가져옵니다.
# - 자연상수(e)는 약 2.71828의 값을 가지며, 수학에서 중요한 상수로 자주 사용됩니다.
from math import e
```

```python
#   자연상수 e의 값을 출력하여 확인합니다.
e
#   출력 결과: 2.718281828459045 (자연상수 e의 근사값입니다.)
```

```python
#   e를 거듭제곱(** 연산자)하여 -1승(역수)을 계산합니다.
#   - e의 -1승은 1/e와 같은 의미로, e의 역수를 나타냅니다.
e ** -1
#   출력 결과: 0.36787944117144233(이는 약 1/e 값입니다.)
```

```python
#   e를 거듭제곱(** 연산자)하여 0승을 계산합니다.
#   - 어던 수라도 0승을 하면 결과는 항상 1입니다.
e ** 0
#   출력 결과: 1.0(거듭제곱에서 0승의 결과는 항상 1입니다.)
```
