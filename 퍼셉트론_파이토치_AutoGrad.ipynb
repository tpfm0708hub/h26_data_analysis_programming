```python
import numpy as np
import torch

from math import e
```

```python
X = np.array([160, 170, 180, 190])

y = np.array([0, 0, 1, 1])
```

```python
X_mean = np.mean(X)

X_mean
```

```python
X_std = np.std(X)

X_std
```

```python
X_norm = (X - X_mean) / X_std

X_norm
```

```python
X_norm_tensor = torch.tensor(X_norm, dtype=torch.float32)

X_norm_tensor
```

```python
y_tensor = torch.tensor(y, dtype=torch.float32)

y_tensor
```

```python
a = torch.tensor(0.1, dtype=torch.float32, requires_grad=True)
a
```

```python
b = torch.tensor(0.0, dtype = torch.float32, requires_grad = True)\
    
b
```

```python
def sigmoid(H):
    return 1 / (1 + e ** (-H) )
```

```python
H = a.item() * X_norm + b.item()

H
```

```python
z = sigmoid(H)

z
```

```python
epsilon = 1e-7

z_safe = np.clip(z, epsilon, 1 - epsilon)

z_safe
```

```python
costs = - ( y * np.log(z_safe) + (1 - y) * np.log(1 - z_safe))

costs
```

```python
mean_costs = np.mean(costs)

mean_costs
```

```python
learning_rate = 0.1

epochs = 1000
```

```python
for epoch in range(epochs):
    H = (a * X_norm_tensor) + b
    z = torch.sigmoid(H)
    
    z_safe = torch.clamp(z, epsilon, 1 - epsilon)
    
    costs = - ( y_tensor * torch.log(z_safe) + (1 - y_tensor) * torch.log( 1- z_safe) )
    mean_cost = torch.mean(costs)
    mean_cost.backward()
    
    grad_a = a.grad.item()
    grad_b = b.grad.item()
    
    with torch.no_grad():
        a -= learning_rate * a.grad
        b -= learning_rate * b.grad
        
    a.grad.zero_()
    b.grad.zero_()
    
    if epoch % 100 == 0 or epoch == epochs -1 :
        print(
            f'Epoch {epoch},'
            f'cost {mean_cost.item()},'
            f'grad_a {grad_a},'
            f'grad_b {grad_b},'
            f'a: {a},'
            f'b: {b}'
            )
```
