```python
model = tf.keras.Sequential([
            tf.keras.Input(shape = (1, )),
            tf.keras.layers.Dense(1)
        ])
```

```python
model.summary()
```

```python
X = np.array([160, 170, 180, 190], dtype = np.float32).reshape(-1, 1)
```

```python
y = np.array([0, 0, 1, 1], dtype = np.float32).reshape(-1, 1)
```

```python
X_mean = X.mean()

X_mean
```

```python
X_std = X.std()

X_std
```

```python
X_norm = (X - X_mean) / X_std

X_norm
```

```python
H_before = model(X_norm)

H_before
```

```python
z_before = tf.sigmoid(H_before)

z_before
```

```python
H_before.numpy()[:3]
```

```python
z_before.numpy()[:2]
```

```python
lr = 0.1
ep = 1000
```

```python
model.compile(
    optimizer = tf.keras.optimizers.SGD(learning_rate = lr) ,
    loss = tf.keras.losses.BinaryCrossentropy(from_logits = True)
)
```

```python
history = model.fit(
                X_norm,
                y,
                epochs = ep,
                verbose = 1
            )
```

```python
history.history['loss']
```

```python
list( enumerate(history.history['loss']) )
```

```python
for epoch_index, mean_cost in enumerate(history.history['loss']):
    if epoch_index % 100 == 0 or epoch_index == ep - 1:
        print(f'epoch = {epoch_index}, mean_cost = {mean_cost}')
```
